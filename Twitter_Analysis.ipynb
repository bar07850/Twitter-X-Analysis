```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.cluster import MiniBatchKMeans
from sklearn.preprocessing import LabelEncoder
import umap
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Limit tweets to reduce memory usage
def sample_df(df, n=10000):
    return df.sample(n=min(n, len(df)), random_state=42)

# Tag and sample each dataset
datasets = {
    'elon': sample_df(elon),
    'kanye': sample_df(kanye),
    'russian_bots': sample_df(russian_bots),
    'obama': sample_df(obama)
}

# Combine into one dataframe with labels
all_data = pd.concat(
    [df.assign(source=name) for name, df in datasets.items()],
    ignore_index=True
)

# Preprocess content (very basic)
all_data['content_clean'] = all_data['content'].str.lower().str.replace(r"http\S+|www.\S+", "", regex=True).str.replace(r"[^a-z\s]", "", regex=True)

# Vectorize
vectorizer = TfidfVectorizer(max_features=3000, stop_words='english')
X = vectorizer.fit_transform(all_data['content_clean'])

# Dimensionality reduction (to speed up clustering and visualization)
svd = TruncatedSVD(n_components=50, random_state=42)
X_reduced = svd.fit_transform(X)

# Clustering
kmeans = MiniBatchKMeans(n_clusters=10, random_state=42)
clusters = kmeans.fit_predict(X_reduced)
all_data['cluster'] = clusters

# 2D projection for visualization
reducer = umap.UMAP(n_components=2, random_state=42)
embedding = reducer.fit_transform(X_reduced)

# Add projection to dataframe
all_data['x'] = embedding[:, 0]
all_data['y'] = embedding[:, 1]

# Plot
plt.figure(figsize=(12, 8))
sns.scatterplot(data=all_data, x='x', y='y', hue='source', style='cluster', palette='tab10', s=50, alpha=0.7)
plt.title("Tweet Content Clusters by Dataset")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def preprocess(series):
    return (
        series
        .dropna()
        .str.lower()
        .str.replace(r"http\S+|www.\S+", "", regex=True)
        .str.replace(r"[^a-z\s]", "", regex=True)
        .str.strip()
        .replace('', np.nan)  # convert empty strings to NaN
        .dropna()  # drop them
    )

# Preprocess and combine datasets
datasets = {
    'elon': preprocess(elon_after['content']),
    'cook': preprocess(cook['content']),
    'bill_gates': preprocess(bill_gates['content']),
    'zuck': preprocess(zuck['content']),
    'jack': preprocess(jack['content']),
    'bezos': preprocess(bezos['content']),
    'russian_bots': preprocess(russian_bots['content']),
}

# Combine all text to fit TF-IDF on shared vocabulary
all_text = pd.concat(datasets.values())

vectorizer = TfidfVectorizer(max_features=3000, stop_words='english')
vectorizer.fit(all_text)

# Compute average vector per dataset
avg_vectors = {}
for name, texts in datasets.items():
    tfidf_matrix = vectorizer.transform(texts)
    avg_vector = np.asarray(tfidf_matrix.mean(axis=0))  # ✅ convert np.matrix to array
    avg_vectors[name] = avg_vector

# Build similarity matrix
names = list(avg_vectors.keys())
matrix = np.vstack([avg_vectors[name] for name in names])
similarity_matrix = cosine_similarity(matrix)

# Convert to DataFrame
sim_df = pd.DataFrame(similarity_matrix, index=names, columns=names)

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(sim_df, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Tweet Content Similarity Between Datasets")
plt.tight_layout()
plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Preprocess function
def preprocess(series):
    return series.str.lower().str.replace(r"http\S+|www.\S+", "", regex=True).str.replace(r"[^a-z\s]", "", regex=True)

# Step 1: Prepare data
elon = elon.sort_values('createdAt')
elon['period'] = elon['createdAt'].dt.to_period('M')  # Group by month

# Preprocess
elon['clean'] = preprocess(elon['content'])
russian_bots['clean'] = preprocess(russian_bots['content'])

# Fit TF-IDF on all data for shared vocabulary
all_text = pd.concat([elon['clean'], russian_bots['clean']])
vectorizer = TfidfVectorizer(max_features=3000, stop_words='english')
vectorizer.fit(all_text)

# Vector for all others (static)
X_bots = vectorizer.transform(russian_bots['clean'])
bots_vector = np.asarray(X_bots.mean(axis=0))

# Step 2: Calculate similarity over time
elon_similarity = []

for period, group in elon.groupby('period'):
    X_elon = vectorizer.transform(group['clean'])
    elon_avg = np.asarray(X_elon.mean(axis=0))
    sim = cosine_similarity(elon_avg, bots_vector)[0][0]
    elon_similarity.append((period.to_timestamp(), sim))


import pandas as pd
import matplotlib.pyplot as plt

# Step 3: Plot
df_sim = pd.DataFrame(elon_similarity, columns=['Date', 'Cosine Similarity'])

plt.figure(figsize=(12, 6))
plt.plot(df_sim['Date'], df_sim['Cosine Similarity'], marker='o', label="Elon vs Russian Bots")

# Define acquisition date
acquisition_date = pd.to_datetime('2022-10-27')

# ✅ Use boolean indexing on the Date column of df_sim
before = df_sim[df_sim['Date'] < acquisition_date]
after = df_sim[df_sim['Date'] >= acquisition_date]

# Plot
plt.figure(figsize=(12, 6))
plt.plot(before['Date'], before['Cosine Similarity'], marker='o', color='gray', label='Before Acquisition')
plt.plot(after['Date'], after['Cosine Similarity'], marker='o', color='red', label='After Acquisition')

# Vertical line for acquisition
plt.axvline(acquisition_date, color='blue', linestyle='--', label='Twitter Acquisition')
plt.axhline(0, color='black', linestyle='--')

plt.title("Similarity of Elon's Tweets to Russian Bots Over Time")
plt.ylabel("Cosine Similarity (0–1)")
plt.xlabel("Time")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Assign dominant topic to each tweet
combined['topic'] = topic_matrix.argmax(axis=1)

topic_counts = combined.groupby(['source', 'topic']).size().unstack(fill_value=0)

# Normalize to percentage per source
topic_dist = topic_counts.div(topic_counts.sum(axis=1), axis=0)


plt.figure(figsize=(12, 6))
topic_ids = topic_dist.columns
x = range(len(topic_ids))

plt.plot(x, topic_dist.loc['elon'], label='Elon Musk')
plt.plot(x, topic_dist.loc['bots'], label='Russian Bots', linestyle='--')
plt.xlabel("Topic #")
plt.ylabel("Topic Proportion")
plt.title("Distribution of Topics in Elon vs Russian Bot Tweets")
plt.legend()
plt.grid(True)
plt.xticks(x)
plt.tight_layout()
plt.show()
```

```python
# Show top words for each topic
def print_topics(model, vectorizer, top_n=10):
    feature_names = vectorizer.get_feature_names_out()
    for idx, topic in enumerate(model.components_):
        top_words = [feature_names[i] for i in topic.argsort()[-top_n:]]
        print(f"Topic {idx}: {' | '.join(top_words)}")

print_topics(lda2, vectorizer, top_n=10)
```

```python
import matplotlib.pyplot as plt

def plot_top_words(lda_model, feature_names, topic_idx, top_n=10):
    topic = lda_model.components_[topic_idx]
    top_indices = topic.argsort()[-top_n:]
    top_features = [feature_names[i] for i in top_indices]
    weights = topic[top_indices]

    plt.figure(figsize=(8, 4))
    plt.barh(top_features, weights)
    plt.title(f"Top words for Topic {topic_idx}")
    plt.xlabel("Weight")
    plt.tight_layout()
    plt.show()

plot_top_words(lda2, vectorizer.get_feature_names_out(), topic_idx=8)
```
